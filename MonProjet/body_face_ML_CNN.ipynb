```python
import pandas as pd
df = pd.read_csv('handsAndFace.csv', sep=";")
```

```python
df_clean = df.fillna(0)
```

```python
X = df_clean.drop('class', axis=1) # features
y = df_clean['class'] # target value
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
from sklearn.preprocessing import LabelEncoder

# Instancier l'encodeur de labels
encoder = LabelEncoder()

# Encoder les étiquettes de classe
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
```

```python
import keras
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Embedding
from keras.utils import to_categorical
from keras.layers import Input, Flatten, Dropout, Activation
from keras.layers import Conv1D, MaxPooling1D
from keras.models import Model
from keras.callbacks import ModelCheckpoint
from keras import optimizers
from keras import layers, models, callbacks
from tensorflow.python.keras import regularizers
# Construction du modèle DCNN
model = models.Sequential([
    layers.Conv1D(64, 5, activation='relu', input_shape=(2172, 1)),
    layers.BatchNormalization(),
    layers.Conv1D(64, 5, activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.MaxPooling1D(2),
    layers.Conv1D(128, 5, activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.Conv1D(128, 5, activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.MaxPooling1D(2),
    layers.Conv1D(256, 5, activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.Conv1D(256, 5, activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.GlobalAveragePooling1D(),
    layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    layers.Dropout(0.5),
    layers.Dense(7, activation='softmax')  # Nombre de classes
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train,y_train_encoded, epochs=10, batch_size=32, validation_data=(X_test, y_test_encoded))
```

```python
loss, accuracy = model.evaluate(X_test, y_test_encoded)
print(f'Accuracy: {accuracy}')
```

```python
y_pred = model.predict(X_test)
```

```python
y_train_pred = model.predict(X_train)
```

```python
from sklearn.metrics import classification_report

# Convertir les prédictions de probabilités en classes prédites (indices)
import numpy as np
y_test_pred_classes = np.argmax(y_pred, axis=1)
y_train_pred_classes = np.argmax(y_train_pred, axis=1)

# Affichage des rapports de classification
print("Rapport de classification pour l'ensemble de test :")
print(classification_report(y_test_encoded, y_test_pred_classes))

print("\nRapport de classification pour l'ensemble d'entraînement :")
print(classification_report(y_train_encoded, y_train_pred_classes))
```

```python
import pandas as pd
df = pd.read_csv('handsAndFace.csv', sep=";")
```

```python
df_clean = df.fillna(0)
```

```python
X = df_clean.drop('class', axis=1) # features
y = df_clean['class'] # target value
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
from sklearn.preprocessing import LabelEncoder

# Instancier l'encodeur de labels
encoder = LabelEncoder()

# Encoder les étiquettes de classe
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
```

```python

from sklearn.svm import SVC

svclassifier = SVC(kernel = 'linear')
```

```python
import time

starting_time = time.time()
print(X_train)
svclassifier.fit(X_train, y_train)
print("Trained model in %s ms " % str(time.time() - starting_time))
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
import seaborn as sn

print(classification_report(y_test,y_pred))

acc = float(accuracy_score(y_test,y_pred))*100
print("----accuracy score %s ----" % acc)

cm = confusion_matrix(y_test,y_pred)
df_cm = pd.DataFrame(cm)
sn.heatmap(df_cm, annot=True, fmt='')
plt.show()
```
