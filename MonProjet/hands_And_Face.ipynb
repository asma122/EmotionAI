```python
from sklearn.pipeline import make_pipeline 
from sklearn.preprocessing import StandardScaler 

from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
pipelines = {
    'lr':make_pipeline(StandardScaler(), LogisticRegression()),
    'rc':make_pipeline(StandardScaler(), RidgeClassifier()),
    'rf':make_pipeline(StandardScaler(), RandomForestClassifier()),
    'gb':make_pipeline(StandardScaler(), GradientBoostingClassifier()),
}
```

```python
fit_models = {}
for algo, pipeline in pipelines.items():
    model = pipeline.fit(X_train, y_train)
    fit_models[algo] = model
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import joblib

# Charger les données
facial_data = pd.read_csv('mergedFace.csv')
hands_data = pd.read_csv('mergedHands.csv')

facial_data = facial_data.fillna(0)
hands_data = hands_data.fillna(0)

# Assurez-vous que les colonnes de classe ont le même nom pour la fusion
facial_data.rename(columns={'class': 'class'}, inplace=True)
hands_data.rename(columns={'class': 'class'}, inplace=True)

# Standardiser les données
scaler_facial = StandardScaler()
scaler_hands = StandardScaler()

facial_features = facial_data.drop(columns=['class'], axis=1)
hands_features = hands_data.drop(columns=['class'], axis=1)

facial_features_scaled = scaler_facial.fit_transform(facial_features)
hands_features_scaled = scaler_hands.fit_transform(hands_features)

# Combiner les caractéristiques et les classes
combined_features = np.hstack((facial_features_scaled, hands_features_scaled))
combined_classes = facial_data['class'].values  # suppose que les classes sont les mêmes dans les deux fichiers

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(combined_features, combined_classes, test_size=0.2, random_state=42)
```

```python
# Former un modèle de RandomForest
model = RandomForestClassifier(n_estimators=7, random_state=42)
model.fit(X_train, y_train)

# Prédire et évaluer le modèle
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
```

```python
import cv2
import mediapipe as mp
import numpy as np
import joblib

# Charger le modèle formé et les scalers
model = joblib.load('model.pkl')
scaler_facial = joblib.load('scaler_facial.pkl')
scaler_hands = joblib.load('scaler_hands.pkl')

# Initialiser MediaPipe
mp_holistic = mp.solutions.holistic
holistic = mp_holistic.Holistic()

# Initialiser OpenCV
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Convertir l'image en RGB
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Faire des prédictions avec MediaPipe
    results = holistic.process(image)
    
    # Extraire les caractéristiques faciales et des mains
    facial_features = []
    hands_features = []
    
    if results.face_landmarks:
        for landmark in results.face_landmarks.landmark:
            facial_features.extend([landmark.x, landmark.y, landmark.z])
    
    if results.left_hand_landmarks:
        for landmark in results.left_hand_landmarks.landmark:
            hands_features.extend([landmark.x, landmark.y, landmark.z])
    
    if results.right_hand_landmarks:
        for landmark in results.right_hand_landmarks.landmark:
            hands_features.extend([landmark.x, landmark.y, landmark.z])
    
    if facial_features and hands_features:
        # Convertir en numpy array et appliquer la standardisation
        facial_features = np.array(facial_features).reshape(1, -1)
        hands_features = np.array(hands_features).reshape(1, -1)
        
        facial_features_scaled = scaler_facial.transform(facial_features)
        hands_features_scaled = scaler_hands.transform(hands_features)
        
        # Combiner les caractéristiques
        combined_features = np.hstack((facial_features_scaled, hands_features_scaled))
        
        # Faire une prédiction avec les probabilités
        probabilities = model.predict_proba(combined_features)[0]
        
        # Afficher les probabilités des classes
        class_labels = model.classes_
        for i, prob in enumerate(probabilities):
            cv2.putText(frame, f'{class_labels[i]}: {prob:.2f}', (10, 30 + i * 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
    
    # Afficher l'image
    cv2.imshow('Real-time Prediction', frame)
    
    # Sortir de la boucle avec 'q'
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import joblib

# Charger les données
data = pd.read_csv('handsAndFace.csv', sep=";")
data = data.dropna() 
# Séparer les caractéristiques et la classe
features = data.drop(columns=['class'], axis=1)
class_labels = data['class'].astype('category')

# Standardiser les données
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(features_scaled, class_labels, test_size=0.2, random_state=42)

# Former un modèle de RandomForest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Prédire et évaluer le modèle
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')

# Sauvegarder le modèle et le scaler
joblib.dump(model,'model.pkl')
joblib.dump(scaler,'scaler.pkl')
```
