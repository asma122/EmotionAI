```python
import pandas as pd
import csv  # Importer le module csv

# Charger le fichier CSV avec des paramètres ajustés
df = pd.read_csv(
    'E:/meldDatabase/MELD-RAW/MELD.Raw/train/train_sent_emo.csv',
    sep=",",                  # Délimiteur (virgule)
    encoding='ISO-8859-1',   # Encodage
    quotechar='"',            # Caractère de citation
    quoting=csv.QUOTE_ALL,    # Citer tous les champs
    skipinitialspace=True,    # Ignorer les espaces après les virgules
    on_bad_lines='skip'       # Ignorer les lignes incorrectes
)

# Afficher les premières lignes du DataFrame
print("Données originales :")
print(df.head())

# Convertir la colonne 'Utterance' en chaînes, en remplaçant les NaN par des chaînes vides
df['Utterance'] = df['Utterance'].astype(str).str.replace('Â', '', regex=False)
df['Utterance'] = df['Utterance'].str.replace('', "'", regex=False)  # Remplacer les caractères indésirables

# Vérifier les noms de colonnes
print("\nNoms des colonnes :")
print(df.columns)

# Afficher les données nettoyées
print("\nDonnées nettoyées :")
print(df.head())

# Enregistrer le fichier nettoyé
df.to_csv('E:/meldDatabase/MELD-RAW/MELD.Raw/train/train_sent_emo_cleaned22.csv', index=False, encoding='ISO-8859-1')

print("\nFichier nettoyé enregistré avec succès.")



```
