```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle

# Charger les fichiers CSV
df_train = pd.read_csv('E:/meldDatabase/MELD-RAW/MELD.Raw/train/train_sent_emo.csv', delimiter=',')
df_test = pd.read_csv('E:/meldDatabase/MELD-RAW/MELD.Raw/test/test_sent_emo.csv', delimiter=',')
df_dev = pd.read_csv('E:/meldDatabase/MELD-RAW/MELD.Raw/dev/dev_sent_emo.csv', delimiter=',')

# Prétraitement des données d'entraînement
texts_train = df_train['Utterance'].values
labels_train = df_train['Emotion'].values

# Encodage des étiquettes
label_encoder = LabelEncoder()
labels_encoded_train = label_encoder.fit_transform(labels_train)

# Tokenisation
tokenizer = Tokenizer(num_words=10000)  # Ajustez selon vos besoins
tokenizer.fit_on_texts(texts_train)
X_train = tokenizer.texts_to_sequences(texts_train)

# Padding des séquences
max_length = 50  # Longueur maximale des séquences
X_train = pad_sequences(X_train, maxlen=max_length)

# Prétraitement des données de test
texts_test = df_test['Utterance'].values
labels_test = df_test['Emotion'].values
labels_encoded_test = label_encoder.transform(labels_test)
X_test = tokenizer.texts_to_sequences(texts_test)
X_test = pad_sequences(X_test, maxlen=max_length)

# Prétraitement des données de validation
texts_dev = df_dev['Utterance'].values
labels_dev = df_dev['Emotion'].values
labels_encoded_dev = label_encoder.transform(labels_dev)
X_dev = tokenizer.texts_to_sequences(texts_dev)
X_dev = pad_sequences(X_dev, maxlen=max_length)

# Construction du modèle RNN
model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=128, input_length=max_length))
model.add(SimpleRNN(64, return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(len(label_encoder.classes_), activation='softmax'))

# Compilation du modèle
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Entraînement du modèle
model.fit(X_train, labels_encoded_train, epochs=10, batch_size=32, validation_data=(X_dev, labels_encoded_dev))

# Évaluation du modèle
train_accuracy = model.evaluate(X_train, labels_encoded_train, verbose=0)[1]
test_accuracy = model.evaluate(X_test, labels_encoded_test, verbose=0)[1]
dev_accuracy = model.evaluate(X_dev, labels_encoded_dev, verbose=0)[1]

print(f"Accuracy on Training Set: {train_accuracy}")
print(f"Accuracy on Test Set: {test_accuracy}")
print(f"Accuracy on Dev Set: {dev_accuracy}")

# Prédictions et rapport de classification
y_pred = model.predict(X_test)
y_pred_classes = tf.argmax(y_pred, axis=1).numpy()
print(classification_report(labels_encoded_test, y_pred_classes, target_names=label_encoder.classes_))

# Sauvegarder le modèle et les outils de prétraitement
model.save('dialogue_rnn_model.h5')
with open('dialogue_rnn_tokenizer.pkl', 'wb') as file:
    pickle.dump(tokenizer, file)
with open('dialogue_rnn_label_encoder.pkl', 'wb') as file:
    pickle.dump(label_encoder, file)

print("Modèle et outils de prétraitement sauvegardés.")
```

```python
import pandas as pd
import re


df = pd.read_csv('E:/meldDatabase/MELD-RAW/MELD.Raw/test/test_sent_emo.csv', delimiter=',')

# Fonction pour nettoyer le texte des caractères spéciaux
def clean_text(text):
    # Remplacer les caractères comme Â par des espaces ou rien
    text = re.sub(r'Â|â€™', "'", text)  # Remplacer les caractères mal encodés
    text = re.sub(r'[^a-zA-Z0-9\s\'\.,?!]', '', text)  # Supprimer les caractères spéciaux indésirables
    text = re.sub(r'\s+', ' ', text).strip()  # Supprimer les espaces multiples
    return text

# Appliquer la fonction à la colonne Utterance
df['Utterance'] = df['Utterance'].apply(clean_text)

# Sauvegarder le fichier nettoyé
output_file = 'E:/meldDatabase/MELD-RAW/MELD.Raw/test/test_sent_emo_cleaned.csv'
df.to_csv(output_file, index=False)

print("Le fichier nettoyé a été sauvegardé sous le nom:", output_file)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, GRU, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import pickle

# Charger les fichiers CSV de données d'entraînement et de test
train_df =  pd.read_csv('E:/meldDatabase/MELD-RAW/MELD.Raw/train/train_sent_emo_cleaned.csv', delimiter=',')

test_df =  pd.read_csv('E:/meldDatabase/MELD-RAW/MELD.Raw/test/test_sent_emo_cleaned.csv', delimiter=',')
# Prétraitement des données d'entraînement
train_texts = train_df['Utterance'].values
train_labels = train_df['Emotion'].values

# Prétraitement des données de test
test_texts = test_df['Utterance'].values
test_labels = test_df['Emotion'].values

# Encodage des étiquettes
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)
test_labels_encoded = label_encoder.transform(test_labels)

# Conversion des étiquettes en catégories (one-hot encoding)
train_labels_categorical = to_categorical(train_labels_encoded)
test_labels_categorical = to_categorical(test_labels_encoded)

# Tokenisation du texte
tokenizer = Tokenizer(num_words=10000)  # Ajustez num_words selon vos besoins
tokenizer.fit_on_texts(train_texts)

# Convertir les textes en séquences
train_sequences = tokenizer.texts_to_sequences(train_texts)
test_sequences = tokenizer.texts_to_sequences(test_texts)

# Padding des séquences pour qu'elles aient la même longueur
X_train = pad_sequences(train_sequences, padding='post')
X_test = pad_sequences(test_sequences, padding='post', maxlen=X_train.shape[1])  # Ajuster la longueur des séquences de test à celles d'entraînement

# Création du modèle
model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=128, input_length=X_train.shape[1]))

# Dropout
model.add(Dropout(0.4))  # Réduit le taux de Dropout

# Ajouter une couche Bidirectionnelle GRU
model.add(Bidirectional(GRU(120, return_sequences=True)))
model.add(Bidirectional(GRU(64, return_sequences=True)))

# Normalisation par lots
model.add(BatchNormalization())

# Ajouter une autre couche Bidirectionnelle GRU
model.add(Bidirectional(GRU(64)))

# Ajouter la couche de sortie
model.add(Dense(7, activation='softmax'))  # 7 classes pour les émotions

# Compilation du modèle avec un taux d'apprentissage spécifique
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Callbacks pour l'arrêt anticipé et l'enregistrement du meilleur modèle
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_loss')

# Entraînement du modèle
model.fit(X_train, train_labels_categorical, epochs=20, batch_size=32, validation_data=(X_test, test_labels_categorical),
          callbacks=[early_stopping, model_checkpoint])

# Évaluation du modèle
loss, accuracy = model.evaluate(X_test, test_labels_categorical)
print(f"Model Accuracy: {accuracy}")

# Sauvegarder le modèle et les outils de prétraitement
model.save('text_emotionMELD_BiGRU_finalTTCleaned.h5')
with open('text_emotionMELD_preprocessingTTCleaned.pkl', 'wb') as file:
    pickle.dump({
        'tokenizer': tokenizer,
        'label_encoder': label_encoder
    }, file)

print("Modèle et outils de prétraitement sauvegardés.")
```
